## ต้องการให้ user, site, ov และ PartitionKey เป็นตัวเลขเท่านั้น

```python
import numpy as np
```

```python
alldb_dp['user'] = pd.to_numeric(alldb_dp['user'], errors='coerce') #errors=’coerce’, you’ll transform the non-numeric values into NaN.
alldb_dp['site'] = pd.to_numeric(alldb_dp['site'], errors='coerce')
alldb_dp['ov'] = pd.to_numeric(alldb_dp['ov'], errors='coerce')
alldb_dp['PartitionKey'] = pd.to_numeric(alldb_dp['PartitionKey'], errors='coerce')
```

```python
alldb_dp = alldb_dp.replace(np.nan, 5, regex=True)
alldb_dp['user'] = alldb_dp['user'].astype(int)
alldb_dp['site'] = alldb_dp['site'].astype(int)
alldb_dp['ov'] = alldb_dp['ov'].astype(int)
alldb_dp['PartitionKey'] = alldb_dp['PartitionKey'].astype(int)
```

## ทำให้ user เป็นตัวเลข 5 หลัก 

```python
TF=list()
for each_row in alldb_dp.iterrows():
    if (9999 < each_row[1]['user'] < 100000):  #each_row คือ ตัวเลข
        TF.append(True)
    else:
        TF.append(False)

alldb_dp1= alldb_dp[TF]
```

```python
alldb_dp1.head()
```

```python
TF2=list()
for each_row in alldb_dp1.iterrows():
    if (9999 < each_row[1]['site'] < 100000):  
        TF2.append(True)
    else:
        TF2.append(False)

alldb_dp2= alldb_dp1[TF2]
```

```python
TF3=list()
for each_row in alldb_dp2.iterrows():
    if (9999 < each_row[1]['PartitionKey'] < 100000):  
        TF3.append(True)
    else:
        TF3.append(False)

alldb_dp3= alldb_dp2[TF3]
```
