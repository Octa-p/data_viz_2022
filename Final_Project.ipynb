### แทน missing ด้วยค่าที่เหมาะสม (fillna())

```python
case = data_cases.fillna(value={'Notified date':'ไม่มีข้อมูล','sex':'ไม่มีข้อมูล','Unit':'ไม่มีข้อมูล','nationality':'ไม่มีข้อมูล','province_of_isolation':'ไม่มีข้อมูล','risk':'ไม่มีข้อมูล','province_of_onset':'ไม่มีข้อมูล','district_of_onset':'ไม่มีข้อมูล'}) #แทนด้วยคำว่า ไม่มีข้อมูล ได้เนื่องจากในตารางมีค่าว่าง และในคอลัมป์ไม่มีค่าเป็นตัวเลขจึงใช้ค่า string ในการแทนที่ได้
```

```python
case
```

### แทนค่า missing ของ age ด้วย mean()

```python
mean_age = case[['age']].mean() # เนื่องจากค่า age เป็นตัวเลขจึงจะต้องแทนค่าด้วยค่าเฉลี่ยของตัวเลขนั้นๆในช่องว่าง
mean_age
```

```python
case[['age']] = case[['age']].fillna(mean_age)
```

```python
case.isnull().any()
```

จากการเตรียมข้อมูลจะเห็นได้ว่า ข้อมูลของจำนวนติดเชื้อช่วงที่ 1 ไม่มีค่า missing แล้ว
